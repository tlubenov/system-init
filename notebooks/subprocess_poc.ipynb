```python
import subprocess
```

```python
process = subprocess.Popen(["bash -i -c  'history -r;history' "], shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
stdout, stderr = process.communicate()
```

```python
for e in stdout.decode("utf-8").split('\n'):
    print(e)
```
